```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # visualization
import seaborn as sns 
# visualization
sns.set(color_codes=True)

# Load the data 
df = pd.read_csv('data.csv')
# To display the top 5 rows
df.head()

```

```python
# Dropping irrelevant columns
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style'], axis=1)
df

```

```python
# Renaming the column names

df = df.rename(columns={'Engine HP' : "HP", "Engine Cylinders":'Cylinders'})
df.head()
```

```python
# Used to count the number of rows before removing the data
df.count()
```

```python
# Dropping the duplicates 
df = df.drop_duplicates()
df.tail()
```

```python
# Counting the number of rows after removing duplicates.
df.count()

```

```python
# Finding the null values.
df.isnull().sum()
```

```python
# Dropping the missing values.
df = df.dropna()
df.count()
```

```python
# After dropping the values
df.isnull().sum()
```

```python
# Detecting Outliers of MSRP
sns.boxplot(x=df['MSRP'])

```

```python
# Detecting Outliers of HP
sns.boxplot(x=df['HP'])

```

```python
# Detecting interquantile range
p1 = df.quantile(0.25)
p2 = df.quantile(0.75)
iqr = p2 - p1
iqr

```

```python
# Removing the outlier
df = df[~((df < (p1-1.5 * iqr)) |(df > (p2 + 1.5 * iqr))).any(axis=1)]
df

```
