```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv('IRIS.csv')

# gaining information from the data
iris.info()
```

```python
# How many flowers for each species are present?
print(iris['species'].value_counts())

# balanced-dataset vs imbalanced datasets
# Iris is a balanced dataset as the number of data points for every class is 50.
# displaying count plot using sns
plt.title("Species Count")
#sns.countplot(iris['species'])

plt.bar(iris['species'],iris['species'].count())
```

```python
#2-D scatter plot:
iris.plot(kind="scatter", x='sepal_length', y='sepal_width')
plt.show()

# What if we color the points by thier class-label/flower-type.
#plt.figure(figsize=(12, 6))
#plt.title("Sepal length and Sepal Width")
#sns.scatterplot(iris['sepal_length'], iris['sepal_width'], hue=iris['species'], s=50)
sns.FacetGrid(iris, hue="species",height=6).map(plt.scatter, "sepal_length", "sepal_width").add_legend()

#plt.show()

# petal length and petal width
#plt.figure(figsize=(12, 6))
#plt.title("Sepal length and Sepal Width")
#plt.scatterplot(iris['petal_length'], iris['petal_width'], hue=iris['species'], s=50)
sns.FacetGrid(iris, hue="species",height=6).map(plt.scatter, "petal_length", "petal_width").add_legend()
plt.show()
```

```python
#Box-plot can be visualized as a PDF on the side-ways.
sns.boxplot(x='species', y='petal_length', data=iris)
plt.show()

```
